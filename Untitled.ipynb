```python
import numpy as np
df['AI_Workload_Ratio'].replace([np.inf,-np.inf],np.nan,inplace=True)
```

```python
df.describe()
```

```python
df=df.drop(columns='Domain')
```

```python
df=df.rename(columns={'Job titiles':'Job title','domain_new':'Domain'})
```

```python
df.shape,df.columns.to_list()
```

```python
df['AI Impact']=df['AI Impact'].apply(lambda x:x.replace('%','')).astype(float)/100
def impact_level(impact):
    if impact >= 0.9:
        return 'Revolutionary'
    elif 0.75 <= impact < 0.9:
        return 'Significant'
    elif 0.5 <= impact < 0.75:
        return 'Moderate'
    elif 0.25 <= impact < 0.5:
        return 'Minor'
    else:
        return 'Negligible'
df['Impact_Level'] = df['AI Impact'].apply(impact_level)
```
